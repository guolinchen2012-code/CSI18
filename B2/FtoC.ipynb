# Import Thu vien

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

# Import du lieu


```python
df=pd.read_csv("ThoiTiet.csv")
df
```

# Chuan hoa

```python
f_input= df["Farenheit"].to_numpy()
c_output= df["Celcius"].to_numpy()
type(f_input), type(c_output)
```

# Train mГҙ hГ¬nh

```python
model=keras.Sequential([
    keras.layers.Dense(units=10, input_shape=[1]),
    keras.layers.Dense(units=1)
])

model.compile(
    loss="mean_squared_error",
    optimizer="adam"
)
model.summary()
```

```python
#train mo hinh
#Epoch: so lan huan luyen tren du lieu 
model.fit(f_input, c_output, epochs=100)
```

# Test mГҙ hГ¬nh

```python
f_user_input= np.array([50])

#thuc te
c_output_real=round((f_user_input[0] - 32) / 1.8,2)
print(f"gia tri C thuc te la:{c_output_real}")

c_output_pred=model.predict(f_user_input)
print(c_output_pred)
print(f"gia tri C du doan:{c_output_pred[0][0]:.2f}C")
```
